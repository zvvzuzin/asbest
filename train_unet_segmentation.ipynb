```python
class AverageMeter(object):
    def __init__(self):
        self.initialized = False
        self.val = None
        self.avg = None
        self.sum = None
        self.count = None

    def initialize(self, val, weight):
        self.val = val
        self.avg = val
        self.sum = val * weight
        self.count = weight
        self.initialized = True

    def update(self, val, weight=1):
        if not self.initialized:
            self.initialize(val, weight)
        else:
            self.add(val, weight)

    def add(self, val, weight):
        self.val = val
        self.sum += val * weight
        self.count += weight
        self.avg = self.sum / self.count

    def value(self):
        return self.val

    def average(self):
        return self.avg
```

```python
from torch import nn
from torch import sigmoid
import segmentation_models_pytorch as smp


device = torch.device("cuda:" + str(torch.cuda.device_count() - 1) if torch.cuda.is_available() else "cpu")

model = smp.Unet(encoder_name='efficientnet-b3', encoder_weights='imagenet', in_channels=1, classes=2,  activation='sigmoid').to(device)

bce = smp.utils.losses.BCEWithLogitsLoss()
dice = smp.utils.losses.DiceLoss()
# criterion = nn.CrossEntropyLoss()
# criterion.__name__= 'loss'


def pixel_acc(pred, label):
        _, preds = torch.max(pred, dim=1)
        valid = (label >= 0).long()
        acc_sum = torch.sum(valid * (preds == label).long())
        pixel_sum = torch.sum(valid)
        acc = acc_sum.float() / (pixel_sum.float() + 1e-10)
        return acc


metrics = [
    smp.utils.metrics.IoU(eps=1.),
    smp.utils.metrics.Fscore(eps=1.),
]


optimizer = torch.optim.Adam([
    {'params': model.decoder.parameters(), 'lr': 1e-5}, 

    {'params': model.encoder.parameters(), 'lr': 1e-5},  
])
```

```python
def save_fig(crop_size, inp_size):
    files = ['asbest/tr_stones/9_12:40:22_05-03-2020_1.png',
             'asbest/tr_stones/1_11:32:12_16-03-2020_1.png',
             'asbest/tr_stones/22_13:21:36_16-03-2020_1.png',
             'asbest/tr_stones/20_12:23:59_16-03-2020_1.png',             
            ]

    full_image = None
    for i, file in enumerate(files):
        img = cv2.imread(file, cv2.IMREAD_UNCHANGED)
        image, st_mask, asb_mask = big_image_predict(model, img, crop_size=crop_size, inp_size=inp_size, device=device)
        if full_image is None:
            full_image = np.concatenate((image, st_mask, asb_mask), axis=0)
        else:
            full_image = np.concatenate((full_image, np.concatenate((image, st_mask, asb_mask), axis=0)), axis=1)
    cv2.imwrite('graphics/' + datetime.now().strftime("%H:%M:%S") + '_segm_images.png', cv2.resize((full_image * 255).astype(np.uint8), (int(full_image.shape[1] / 8), int(full_image.shape[0] / 8))))
#     return full_image
```

```python
from tqdm import trange, tqdm
from torch.utils.data import DataLoader

img_size=(1024, 1024)
crop_size=(1024, 1024)

np.random.seed(17)
np.random.shuffle(anno_tr_stones)

tr_stones_train_data = Asbest_segmentation(anno_tr_stones[:100], 
                                           crop_size=crop_size,#(img_tr_stones_shape[0] // 2, img_tr_stones_shape[1] // 2),
                                           img_size=img_size, 
                                           num_frames=500, 
                                           normalize=True)
tr_stones_valid_data = Asbest_segmentation(anno_tr_stones[100:120], 
                                           crop_size=crop_size,#(img_tr_stones_shape[0] // 2, img_tr_stones_shape[1] // 2), 
                                           img_size=img_size, 
                                           num_frames=100, 
                                           normalize=True)

tr_stones_train_loader = DataLoader(tr_stones_train_data, batch_size=1, shuffle=True, num_workers=8)
tr_stones_valid_loader = DataLoader(tr_stones_valid_data, batch_size=1, shuffle=False, num_workers=2)

epochs = 100

segm_stones_dict = {'loss':[], 'val_loss':[], 'st_iou':[], 'val_st_iou':[], 'st_dice':[], 'val_st_dice':[]}

with tqdm(total=len(tr_stones_train_loader) + len(tr_stones_valid_loader), 
              bar_format='{desc} epoch {postfix[0]} ' + 
              '| {n_fmt}/{total_fmt} {elapsed}<{remaining} ' + 
              '| loss : {postfix[1]:>2.4f} ' + 
#               '| acc : {postfix[2]:>2.4f} ' + 
              '| st_iou: {postfix[2]:>2.4f} ' + 
#               '| asb_iou: {postfix[3]:>2.4f} ' +
              '| st_dice: {postfix[3]:>2.4f} ' + 
              '| val_loss : {postfix[4]:>2.4f} ' + 
#               '| val_acc : {postfix[6]:>2.4f} ' + 
              '| val_st_iou: {postfix[5]:>2.4f} ' + 
#               '| val_asb_iou: {postfix[6]:>2.4f} '
              '| val_st_dice: {postfix[6]:>2.4f}'
              , 
              postfix=[0, 0, 0, 0, 0, 0, 0], desc = 'Training', leave=True) as t:
    for epoch in range(1, epochs +1):

        average_total_loss = AverageMeter()
        average_st_iou = AverageMeter()
        average_st_dice = AverageMeter()

        model.train()

        t.postfix[0] = epoch
        for data in tr_stones_train_loader:

            torch.cuda.empty_cache()
            inputs, st_masks, asb_masks = data

            inputs=inputs.to(device).float()
            st_masks=st_masks.to(device).float()
            asb_masks=asb_masks.to(device).float()

            optimizer.zero_grad()
            outputs = model(inputs)

            loss = 0.5 * bce(outputs[:,0:1,:,:], st_masks) + 0.5 * dice(outputs[:,0:1,:,:], st_masks)# + 0.0 * dice(outputs[:,1:2,:,:], asb_masks)
            st_iou = metrics[0](outputs[:,0:1,:,:], st_masks)
            st_dice = metrics[1](outputs[:,0:1,:,:], st_masks)

            with amp.scale_loss(loss, optimizer) as scaled_loss:
                scaled_loss.backward()
            optimizer.step()

            average_total_loss.update(loss.data.item())
            average_st_iou.update(st_iou.data.item())
            average_st_dice.update(st_dice.data.item())

            t.postfix[1] = average_total_loss.average()
            t.postfix[2] = average_st_iou.average()
            t.postfix[3] = average_st_dice.average()
            t.update()
        
        del inputs, st_masks, asb_masks
        
        ## Validation
        val_average_total_loss = AverageMeter()
        val_average_st_iou = AverageMeter()
        val_average_st_dice = AverageMeter()

        model.eval()
        
        for data in tr_stones_valid_loader:

            torch.cuda.empty_cache()
            inputs, st_masks, asb_masks = data

            inputs=inputs.to(device).float()
            st_masks=st_masks.to(device).float()
            asb_masks=asb_masks.to(device).float()

            optimizer.zero_grad()
            outputs = model(inputs)

            loss = 0.5 * bce(outputs[:,0:1,:,:], st_masks) + 0.5 * dice(outputs[:,0:1,:,:], st_masks) #+ 0.0 * dice(outputs[:,1:2,:,:], asb_masks) + 0.0 * bce(outputs, torch.cat((st_masks, asb_masks),dim=1))
            st_iou = metrics[0](outputs[:,0:1,:,:], st_masks)
            st_dice = metrics[1](outputs[:,0:1,:,:], st_masks)

            val_average_total_loss.update(loss.data.item())
            val_average_st_iou.update(st_iou.data.item())
            val_average_st_dice.update(st_dice.data.item())

            t.postfix[4] = val_average_total_loss.average()
            t.postfix[5] = val_average_st_iou.average()
            t.postfix[6] = val_average_st_dice.average()
            t.update()
            
        segm_stones_dict['loss'].append(average_total_loss.average())
        segm_stones_dict['val_loss'].append(val_average_total_loss.average())
        segm_stones_dict['st_iou'].append(average_st_iou.average())
        segm_stones_dict['val_st_iou'].append(val_average_st_iou.average())
        segm_stones_dict['st_dice'].append(average_st_dice.average())
        segm_stones_dict['val_st_dice'].append(val_average_st_dice.average())
        
        if epoch % 10 == 0:
            save_fig(crop_size=crop_size, inp_size=img_size) #(img_tr_stones_shape[0] // 2, img_tr_stones_shape[1] // 2)
        t.reset()
```

```python
from torch import nn
from torch import sigmoid
import segmentation_models_pytorch as smp


device = torch.device("cuda:" + str(torch.cuda.device_count() - 1) if torch.cuda.is_available() else "cpu")

model = smp.Unet(encoder_name='efficientnet-b3', in_channels=1, activation='sigmoid').to(device)

bce = smp.utils.losses.BCEWithLogitsLoss()
dice = smp.utils.losses.DiceLoss(activation='sigmoid')
# criterion = nn.CrossEntropyLoss()
# criterion.__name__= 'loss'


def pixel_acc(pred, label):
        _, preds = torch.max(pred, dim=1)
        valid = (label >= 0).long()
        acc_sum = torch.sum(valid * (preds == label).long())
        pixel_sum = torch.sum(valid)
        acc = acc_sum.float() / (pixel_sum.float() + 1e-10)
        return acc


metrics = [
    smp.utils.metrics.IoU(eps=1.),
    smp.utils.metrics.Fscore(eps=1.),
]


optimizer = torch.optim.Adam([
    {'params': model.decoder.parameters(), 'lr': 1e-4}, 

    {'params': model.encoder.parameters(), 'lr': 1e-4},  
])
```
